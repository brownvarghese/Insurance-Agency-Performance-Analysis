```python
# We only work on 23 Columns so a data field on those columns

Updated_Ins_df = Ins_df[["AGENCY_ID","PRIMARY_AGENCY_ID","PROD_ABBR","PROD_LINE","STATE_ABBR",
                         "STAT_PROFILE_DATE_YEAR","NB_WRTN_PREM_AMT","WRTN_PREM_AMT",                         
                         "LOSS_RATIO_3YR","GROWTH_RATE_3YR","ACTIVE_PRODUCERS","MAX_AGE","MIN_AGE","VENDOR_IND"]]

Updated_Ins_df.head()

#Updated_Ins_df.to_csv('output.csv', index=False)
```

```python
# Find out the median value of LOSS_RATIO_3YR 
new_df.LOSS_RATIO_3YR.median()
```

```python
# replace the value 99999 with the median value for simplify

new_df.loc[:, 'LOSS_RATIO_3YR'].replace([99999], [new_df.LOSS_RATIO_3YR.median()], inplace=True)
new_df.head()
```

```python
# Drop ACTIVE_PRODUCERS , MAX_AGE & MIN_AGE is 99999

new_df.drop(new_df.loc[new_df['ACTIVE_PRODUCERS']==99999].index, inplace=True)
new_df.drop(new_df.loc[new_df['MAX_AGE']==99999].index, inplace=True)
new_df.drop(new_df.loc[new_df['MIN_AGE']==99999].index, inplace=True)

new_df.head()
```

```python
new_df['AVG_AGE'] = (new_df["MAX_AGE"] - new_df["MIN_AGE"]) / new_df["ACTIVE_PRODUCERS"]

new_df['AGENCY_PERFORMANCE'] = new_df["WRTN_PREM_AMT"]*(1-new_df["LOSS_RATIO_3YR"])

new_df["DEMOGRAPHY"] = ""

new_df["INSURABLE_POPULATION"] = ""
new_df.head()
```

```python
new1_df=new_df.sort_values(['STATE_ABBR', 'STAT_PROFILE_DATE_YEAR'], ascending=[True, True])
new1_df=new1_df.rename(columns = {'STATE_ABBR':'State'})
new1_df=new1_df.rename(columns = {'STAT_PROFILE_DATE_YEAR':'Year'})
new1_df.head()
```

```python
census_df = pd.read_csv("census_ext.csv")
census_df.head()
```

```python
final_agency_df = pd.merge(census_df, new1_df, how='left', on=['State', 'Year'])
final_agency_df
```
